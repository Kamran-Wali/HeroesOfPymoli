### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
#purchase_data.describe()
#purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#purchase_data["SN"].nunique()
totalPlayers = pd.DataFrame([{"Player Count":purchase_data["SN"].nunique()}])
totalPlayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uIdemID = purchase_data["Item ID"].nunique()
uIdemID
avgPPrice = purchase_data["Price"].mean()
avgPPrice
totalPCount = purchase_data["Price"].count()
totalPCount
totalRevenue = purchase_data["Price"].sum()
totalRevenue
priceSummary = pd.DataFrame([{"Number of Unique Items":uIdemID, "Average Price":"$ "+str(round(avgPPrice,2)), "Number of Purchases":totalPCount, "Total Revenue":"$"+str(round(totalRevenue,2))}])
priceSummary = priceSummary[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
priceSummary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
totalGCount = purchase_data["Gender"].count()
groupByGCount = purchase_data["Gender"].value_counts()
groupByGCount
groupbyGpercent = round(groupByGCount / totalGCount*100, 2 )
groupbyGpercent
#Summary of the gender count 
GenderSummary = pd.DataFrame({"Percentage of players":groupbyGpercent,"Total Count":groupByGCount})
GenderSummary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a dataframe grouped by Gender
PurchaseByGender = purchase_data.groupby(["Gender"])
#PurchaseByGender.head()
#Purchase Count
PGCount = PurchaseByGender["Price"].count()
#PGCount
#Average Purchase Price
APPrice = round(PurchaseByGender["Price"].mean(), 2)
#APPrice
#Total Purchase Value
TPValue = round(PurchaseByGender["Price"].sum(), 2)
TPValue
#Average Purchase Total per Person by Gender Normalized
APTperGender = round(TPValue / PGCount, 2)
APTperGender

#Purchase By Gender Summary

PGSummary = pd.DataFrame({"Purchase Count":PGCount,"Average Purchase Price":APPrice,"Total Purchase Value":TPValue,"Avg Purchase Total per Person":APTperGender})
PGSummary = PGSummary[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Purchase Total per Person"]]
PGSummary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins,labels=group_names)
PDGByAge = purchase_data.groupby(['Age Group'])

#PDGByAge.head()

PDGByAgeTC = purchase_data["Age"].count()
#print(PDGByAgeTC)
PDGByAgeVC = purchase_data["Age Group"].value_counts()
#PDGByAgeVC

PDGByAgePC = round(PDGByAgeVC / PDGByAgeTC * 100, 2)

PDGByAgeDF = pd.DataFrame({"Percentage of Players":PDGByAgePC,"Total Count":PDGByAgeTC})
PDGByAgeDF = PDGByAgeDF.sort_index()
PDGByAgeDF
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pdataAge = purchase_data.groupby(["Age Group"])
#purchase_dataAge.head()
#Purchase Count - grouped by age
pdataAgeC = pdataAge["Price"].count()
#pdataAgeC
#Average Purchase Price
pdataAgeA = round(pdataAge["Price"].mean(), 2)
#pdataAgeA
#Total Purchase Value
pdataAgeT = round(pdataAge["Price"].sum(), 2)
#pdataAgeA
#Average Purchase Total per Person
pdataAgeAP = round(pdataAgeA/pdataAgeC*100,2)

#Average Purchase Total per Person by Age Group
pdataAgeS = pd.DataFrame({"Purchase Count":pdataAgeC,"Average Purchase Price":pdataAgeA,"Total Purchase Value":pdataAgeT,"Average Purchase Total per Person":pdataAgeAP})
pdataAgeS = pdataAgeS[["Purchase Count","Average Purchase Price","Total Purchase Value","Average Purchase Total per Person"]]
pdataAgeS
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# New dataframe - Group by SN
pdataSN = purchase_data.groupby("SN")
#pdataSN.head()

# Purchase Count 
pdataSNC = pdataSN["Price"].count()
#pdataSNC

# Average Purchase Price 
pdataSNA = round(pdataSN["Price"].mean(), 2)
#pdataSNA

#Total Purchase Value
pdataSNT = round(pdataSN["Price"].sum(), 2)
#pdataSNT

# Summary of the top 5 spenders in the game by total purchase value, then list
pdataSNS = pd.DataFrame({"Purchase Count":pdataSNC,"Average Purchase Price":pdataSNA,"Total Purchase Value":pdataSNT})
pdataSNS = pdataSNS[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
pdataSNS = pdataSNS.sort_values('Total Purchase Value',ascending=False)
pdataSNSTopFive = pdataSNS.head(5)
pdataSNSTopFive
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item ID
pdataID = purchase_data.groupby(['Item ID','Item Name'])

#Total Purchase Value by ITem
TPbyI =pdataID['Price'].sum()
#TPbyS = TPbyA.sort_values('Price', ascending=False)
#Purchase Count
PCbyI =pdataID['Price'].count()
#Average Purchase Price
APbyI =pdataID['Price'].mean()
#Item Price
IPbyI =pdataID['Price']

TopI = pd.DataFrame({"Purchase Count":PCbyI,"Item Price":APbyI,"Total Purchase Value":TPbyI})
TopI = TopI[["Purchase Count","Item Price","Total Purchase Value"]]
TopI = TopI.sort_values("Purchase Count",ascending=False)
TopI.head()
```
